```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Comparing how the number of rooms affects price
avg_price = df.groupby('Rooms')['Price'].mean().sort_index()

avg_price.plot(kind='bar', figsize=(8,5))
plt.ylabel("Average Price (AUD)")
plt.xlabel("Number of Rooms")
plt.title("Average House Price by Room Count")
plt.show()
```

```python
# Showing whether bigger land areas mean higher property prices
plt.scatter(df['Landsize'], df['Price'], alpha=0.3)
plt.xlabel("Land Size (mÂ²)")
plt.ylabel("Price (AUD)")
plt.title("Price vs Land Size")
plt.xlim(0, 2000)   # limiting extreme outliers for readability
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=50, kde=True)
plt.title("Distribution of House Prices")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Rooms", y="Price", data=df)
plt.title("House Price by Number of Rooms")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Regionname", data=df, order=df["Regionname"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Number of Houses by Region")
plt.show()
```
